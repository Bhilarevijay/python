```python
##########PROJECT###########

import matplotlib.pyplot as plt

# Sample data
file_sizes = ['10MB', '10MB', '100MB', '100MB']
chunk_sizes = ['512KB', '1MB', '512KB', '1MB']
encryption_times = [2.1, 1.8, 18.5, 15.2]

plt.figure(figsize=(10, 6))
plt.bar([f"{fs}\n{cs}" for fs, cs in zip(file_sizes, chunk_sizes)], encryption_times, color='skyblue')
plt.xlabel('File Size & Chunk Size')
plt.ylabel('Encryption Time (s)')
plt.title('Media Encryption Speed vs. Chunk Size')
plt.show()
```
